# K Means Cluster Creation

Now It is time to create the Cluster labels!

**Import KMeans from SciKitLearn**

```python
from sklearn.cluster import KMeans
```

**Create an instance of a K Means model with 2 clusters**

```python
kmeans = KMeans(n_clusters=2)
```

**Fit the model to all the data except for the Private label**

```python
kmeans.fit(data.drop("Private", axis=1))
```

**What are the cluster center vectors?**

```python
kmeans.cluster_centers_
```

```python
kmeans.cluster_centers_.shapeape
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our cluster.

**Create a new column for data called "Cluster", which is a 1 for a Private school and a 0 for a public school.**

```python
def converter(private):
    if private=="Yes":
        return 1
    else:
        return 0 
```

```python
data["Cluster"] = data["Private"].apply(converter)
```

```python
data["Cluster"].value_counts
```

```python
data.head(2)
```

**Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(data["Cluster"],kmeans.labels_))
print("\n")
print(classification_report(data["Cluster"],kmeans.labels_))
```
