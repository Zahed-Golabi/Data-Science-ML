<h2> Data </h2>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

<h2> Exercise 1 </h2>

Follow along with these steps:<br>
- <b>Create a figure object called fig</b>
- <b>Use add_axes to add an axis to the figure canvas at [0,0,1,1]</b>
- <b>Plot (x,y) on that axes and set the labels and titles</b>

```python
# Create a figure object
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title("title")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

<h2> Exercise 2 </h2>

<b> Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,0.2,0.2]

```python
# Create figure object
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

<b> Now plot(x,y) on both axes And call your figure object to show it</b>

```python
ax1.plot(x,y)
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.plot(x,y)
ax2.set_xlabel("x")
ax2.set_ylabel("y")
```

<h2> Exercise 3 </h2>

<b> Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,0.4]

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])

ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

<b> Now use x,y and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:

```python
# Large plot
ax1.plot(x,z, color="blue")
ax1.plot(x,y, color="blue")
ax1.set_xlabel("x")
ax1.set_ylabel("z")

# Inserted plot
ax2.set_title("zoom")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_xlim(xmin=20.0, xmax=22.0)
ax2.set_ylim(ymin=30, ymax=50)
ax2.plot(x,y, color="red")
```

<h2> Exercise 4</h2>

<b> Use plt.subplots(nrows=1, ncols=2) to create the plot below.</b>

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=())
```

<b> Now plot (x,y) and (x,z) on the axes. Please around with the linewidth and style.</b>

```python
ax1, ax2 = ax
```

```python
ax1.plot(x,y, color="blue", lw=2, ls="--")
ax2.plot(x,z, color="red", lw=3)
```
