```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({"a":list(np.random.rand(500)), "b":list(np.random.rand(500)),
                     "c":list(np.random.rand(500)), "d":list(np.random.rand(500))})
```

<b> Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. </b>

```python
data.plot.scatter(x="a", y="b", s=100, color="red", figsize=(12,3))
```

<b> Create a histogram of the "a" column </b>

```python
plt.style.use("ggplot")
data["a"].plot.hist(bins=15, alpha=0.5)
```

<b> These plots are okay, but they don't look very polished. Use style sheets to set the style to "ggplot" and redo the histogram from above.</b>

```python
sns.set_style(style="whitegrid")
data["a"].plot.hist(bins=30, color="purple")
```

<b> Create a boxplot comparing the a and b columns. </b>

```python
data[["a","b"]].plot.box()
```

<b> Create a kde plot of the "d" column </b>

```python
data["d"].plot.kde()
```

<b> Figure out how to increase the linewidth and make the linestyle dashed. (Note: You would usually not dash a kde plot line)</b>

```python
  data["d"].plot.kde(lw=3, ls="--", color="orange")
```

<b> Create an area plot of all the columns for just the rows up to 30. (hint:use .ix)</b>

```python
data.plot.area(stacked=True)
```

```python
data.iloc[0:30].plot.area(alpha=0.5)
```

<b> Figure out how to display the legend outside of the plot </b>

```python
data.iloc[0:30].plot.area(alpha=0.5)
plt.legend(loc="center left", bbox_to_anchor=(1.0,0.5))
```
