```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head(7)
```

```python
sns.jointplot(x="fare", y="age", data=titanic)
```

```python
sns.jointplot(x="fare", y="age", data=titanic, hue="survived")
```

```python
sns.barplot(x="sex", y="survived", data=titanic)
```

```python
sns.barplot(x="sex", y="fare", data=titanic)
```

```python
sns.histplot(titanic["fare"], color="red", bins=30)
```

```python
sns.boxplot(x="class", y="age", data=titanic, hue="sex")
```

```python
#sns.pointplot(x="class", y="age", data=titanic)
sns.swarmplot(x="class", y="age", data=titanic, palette="Set2")
```

```python
sns.countplot(x="sex", data=titanic)
```

```python
sns.heatmap(titanic.corr(), cmap="coolwarm")
plt.title("titanic.corr()")
```

```python
fg = sns.FacetGrid(data=titanic, col="sex")
fg.map(plt.hist, "age")
```
