<h2> GroupBy </h2>

- Groupby allows you to group together rows based off of a column and 
perform and an aggregate function on them

<img src="Groupby.png">

```python
data = {"Company":["Goog","Goog","MSFT","MSFT","FB","FB"],
       "Person":["Sam","Charlie","Amy","Vanessa","Carl","Sarah"],
       "Sales":[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby("Company")
```

```python
byComp = df.groupby("Company")
```

```python
byComp.mean()
```

```python
byComp.sum()
```

```python
byComp.std()
```

```python
byComp.std().loc["FB"]
```

```python
df.groupby("Company").sum().loc["FB"]
```

```python
df.groupby("Company").count()
```

```python
df.groupby("Company").min()
```

```python
df.groupby("Company").max()
```

```python
df.groupby("Company").describe()
```

```python
df.groupby("Company").describe().transpose()
```

```python
df.groupby("Company").describe().transpose()["FB"]
```

<h2> Merging, Joining, and Concatenating </h2>

- There are 3 main ways of combining DataFrames together:
    1. Mergin
    2. Joining
    3. Concatenating


```python
df1 = pd.DataFrame({"A":["A0","A1","A2","A3"],
                   "B":["B0","B1","B2","B3"],
                   "C":["C0","C1","C2","C3"],
                   "D":["D0","D1","D2","D3"]}, index=[0,1,2,3])
```

```python
df2 = pd.DataFrame({"A":["A4","A5","A6","A7"],
                   "B":["B4","B5","B6","B7"],
                   "C":["C4","C5","C6","C7"],
                   "D":["D4","D5","D6","D7"]}, index=[4,5,6,7])
```

```python
df3 = pd.DataFrame({"A":["A8","A9","A10","A11"],
                   "B":["B8","B9","B10","B11"],
                   "C":["C8","C9","C10","C11"],
                   "D":["D8","D9","D10","D11"]}, index=[8,9,10,11])
```

<h2> Merging </h2>

The merge function allows you to merge DataFrames together using a similar logic as merging SQL Tables together.
For Example:

```python
left = pd.DataFrame({"key":["k0","k1","k2","k3"],
                    "A":["A0","A1","A2","A3"],
                    "B":["B0","B1","B2","B3"]})


right = pd.DataFrame({"key":["k0","k1","k2","k3"],
                     "C":["C0","C1","C2","C3"],
                     "D":["D0","D1","D2","D3"]})
                    
```

```python
left
```

```python
right
```

```python
pd.merge(left,right, how="inner", on="key")
```

<h3> More Complicated examples:</h3>

```python
left = pd.DataFrame({"key1":["k0","k0","k1","k2"],
                    "key2":["k0","k1","k0","k1"],
                    "A":["A0","A1","A2","A3"],
                    "B":["B0","B1","B2","B3"]})

right = pd.DataFrame({"key1":["k0","k1","k1","k2"],
                    "key2":["k0","k0","k0","k0"],
                    "C":["C0","C1","C2","C3"],
                    "D":["D0","D1","D2","D3"]})
```

```python
pd.merge(left,right,on=["key1","key2"])
```

```python
pd.merge(left, right, how="outer", on=["key1","key2"])
```

```python
 pd.merge(left, right, how="right", on=["key1","key2"])
```

```python
 pd.merge(left, right, how="left", on=["key1","key2"])
```

<h2> Joining </h2>

Joining is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame. 

```python
left = pd.DataFrame({"A":["A0","A1","A2"],
                    "B":["B0","B1","B2"]
                    }, index=["k0","k1","k2"])

right = pd.DataFrame({"C":["C0","C2","C3"],
                     "D":["D0","D2","D3"]},
                    index=["k0","k2","k3"])
```

```python
left.join(right)
```

```python
left.join(right, how="outer")
```
